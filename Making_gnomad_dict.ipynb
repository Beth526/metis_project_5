```python
gnomad_dict = defaultdict(list)

for i in gnomad_files:
    file_name = i[2:]
    current_file = pd.read_csv(file_name, sep=" ", header=None)
    
    #change insertion to the same format as TCGA MAF ("-" "ATG" where "A" is not part of ref)
    current_file.loc[current_file[3].apply(len) > 1, 2] = "-"
    current_file.loc[current_file[3].apply(len) > 1, 3] = current_file[3].str.slice(start=1)
    
    #change deletion to the same format as TCGA MAF ("ATG" "-")
    current_file.loc[current_file[2].apply(len) > 1, 2] = current_file[2].str.slice(start=1)
    current_file.loc[current_file[2].apply(len) > 1, 3] = "-"

    for i in range(len(current_file)):
        gnomad_dict[current_file.loc[i,0]].append([current_file.loc[i,1], current_file.loc[i,2], current_file.loc[i,3], current_file.loc[i, 4]])
```

```python
import pickle
```

```python
with open('gnomad_dictionary_2.pickle', 'wb') as write_file:
    pickle.dump(gnomad_dict, write_file)
```
