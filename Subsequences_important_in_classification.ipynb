# Subsequences important in classification

These were the subsequences from the 1000 sequences that the model scored most high in each tumor category from the combined train and validation datasets. The sequences are not necessarily correctly classified, but they represent what the model thought of as most strongly belonging to each class. 

The data requires some more summarization, because each dataframe has around 2000 subsequences in it, to identify commonalities among the subsequences

### Index part 0 is the reference (dashes are insertions), Index part 1 is the altered sequence (dashes are deletions)
